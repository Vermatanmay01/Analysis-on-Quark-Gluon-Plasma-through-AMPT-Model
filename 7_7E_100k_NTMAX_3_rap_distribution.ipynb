```python
import ROOT

tfile1 = ROOT.TFile('Events_at_7_7E_NTMAX_3_100000events.root', 'r')
plot1 = tfile1.Get('test')

h1_piplus = ROOT.TH1F("Pion_rap_Spectra","Pion rap Spectra",18,-4,4)
H1_piminus = ROOT.TH1F("Antipion_rap_Spectra","Antipion rap Spectra",18,-4,4)
k1_kplus = ROOT.TH1F("kaon_rap_Spectra","kaon rap Spectra",18,-4,4)
K1_kminus  = ROOT.TH1F("Antikaon_rap_Spectra","Antikaon rap Spectra",18,-4,4)
p1_Pplus = ROOT.TH1F("proton_rap_Spectra","proton rap Spectra",18,-4,4)
P1_Pminus  = ROOT.TH1F("Antiproton_rap_Spectra","Antiproton rap Spectra",18,-4,4)
l1_Lplus = ROOT.TH1F("lambda_rap_Spectra","lambda rap Spectra",18,-4,4)
L1_Lminus  = ROOT.TH1F("Antilambda_rap_Spectra","Antilambda rap Spectra",18,-4,4)
s1_Splus = ROOT.TH1F("sigma_rap_Spectra","sigma rap Spectra",18,-4,4)
S1_Sminus  = ROOT.TH1F("Antisigma_rap_Spectra","Antisigma rap Spectra",18,-4,4)
k01_k_s  = ROOT.TH1F("kaon_zero_short_rap_Spectra","kaon zero short rap Spectra",18,-4,4)
k01_k_l  = ROOT.TH1F("kaon_zero_long_rap_Spectra","kaon zero long rap Spectra",18,-4,4)
omega_plus = ROOT.TH1F("omega_rap_Spectra","omega rap Spectra",18,-4,4)
# eta_plus = ROOT.TH1F("eta_rap_Spectra","eta rap Spectra",9,-4,4)
```

```python
num_bins = 18
rapidity_min = -4.0
rapidity_max = 4.0
bin_width = (rapidity_max - rapidity_min) / num_bins

for entry in plot1:
    for i in range(entry.pion_count):
        h1_piplus.Fill(entry.pion_rap[i])
    for i in range(entry.antipion_count):
        H1_piminus.Fill(entry.antipion_rap[i])
    for i in range(entry.kaon_count):
        k1_kplus.Fill(entry.kaon_rap[i])
    for i in range(entry.antikaon_count):
        K1_kminus.Fill(entry.antikaon_rap[i])
    for i in range(entry.proton_count):
        p1_Pplus.Fill(entry.proton_rap[i])
    for i in range(entry.antiproton_count):
        P1_Pminus.Fill(entry.antiproton_rap[i])
    for i in range(entry.lambda_count):
        l1_Lplus.Fill(entry.lambda_rap[i])
    for i in range(entry.antilambda_count):
        L1_Lminus.Fill(entry.antilambda_rap[i])
    for i in range(entry.sigma_count):
        s1_Splus.Fill(entry.sigma_rap[i])
    for i in range(entry.antisigma_count):
        S1_Sminus.Fill(entry.antisigma_rap[i])
    for i in range(entry.kaon_zero_short_count):
        k01_k_s.Fill(entry.kaon_zero_short_rap[i])
    for i in range(entry.kaon_zero_long_count):
        k01_k_l.Fill(entry.kaon_zero_long_rap[i])
    for i in range(entry.omega_count):
        omega_plus.Fill(entry.omega_rap[i])
        
#Scaling these histograms
h1_piplus.Scale(1.0 / (bin_width * plot1.GetEntries()))
H1_piminus.Scale(1.0 / (bin_width * plot1.GetEntries()))
k1_kplus.Scale(1.0 / (bin_width * plot1.GetEntries()))  
K1_kminus.Scale(1.0 / (bin_width * plot1.GetEntries()))    
p1_Pplus.Scale(1.0 / (bin_width * plot1.GetEntries()))    
P1_Pminus.Scale(1.0 / (bin_width * plot1.GetEntries()))
l1_Lplus.Scale(1.0 / (bin_width * plot1.GetEntries()))
L1_Lminus.Scale(1.0 / (bin_width * plot1.GetEntries()))
s1_Splus.Scale(1.0 / (bin_width * plot1.GetEntries()))
S1_Sminus.Scale(1.0 / (bin_width * plot1.GetEntries()))
k01_k_s.Scale(1.0 / (bin_width * plot1.GetEntries()))
k01_k_l.Scale(1.0 / (bin_width * plot1.GetEntries()))
```

```python
g1 = ROOT.TCanvas()

h1_piplus.SetLineColor(2) #red
h1_piplus.Draw()


h1_piplus.GetYaxis().SetTitle("dn/dy")
h1_piplus.GetXaxis().SetTitle("y")
# h1_piplus.SetTitle("Transverse momentum distribution")
h1_piplus.SetMarkerStyle(21)

# legend = ROOT.TLegend(0.7,0.6,0.8,0.9)
# legend.AddEntry(h1_piplus,"0-5%")

# legend.Draw("same")

# g1.SetLogy()
g1.Draw()
```
